### Importar bibliotecas

```python
import pandas as pd
import numpy as np
from pycaret.classification import *
from pycaret.datasets import get_data
```

### Carregar Dados

```python
df = get_data('juice')
print(df.shape)
```

### Preparar o setup do Pycaret

```python
setup1= setup(data = df,
              target = 'Purchase',
              normalize= True,
              normalize_method= 'minmax',
              silent= True

              )
```

```python
get_config(variable= "X_train")
```

### Treinar diferentes modelos

```python
compare_models( sort= 'AUC')
```

```python
model = create_model('lr')
```

```python
tuned_model= tune_model(model, n_iter = 100, choose_better= True)
```

```python
evaluate_model(model)
```

### Salvar o modelo para o deploy

```python
finalize_model(tuned_model)
save_model(tuned_model,'final_model')
```
