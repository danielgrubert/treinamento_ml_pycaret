### Importar bibliotecas

```python
import pandas as pd
import numpy as np
from pycaret.regression import *
from pycaret.datasets import get_data
```

### Carregar Dados

```python
data = get_data('boston')
print(data.shape)
```

### Preparar o setup do Pycaret

```python
setup1= setup(data = data, target = 'medv', session_id= 123, pca= True, silent= True, normalize= True, normalize_method= "minmax")
```

```python
get_config(variable= "X_train")
```

### Treinar diferentes modelos

```python
compare_models()
```

```python
model = create_model('et')
```

```python
tuned_model= tune_model(model, n_iter = 250, choose_better= True)
```

```python
evaluate_model(model)
```

### Salvar o modelo para o deploy

```python
finalize_model(tuned_model)
save_model(tuned_model,'final_model')
```
